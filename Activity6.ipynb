```python
plt.scatter(df['Position'],df['Age'])
plt.title('scatter plot for Position and Age')
plt.xlabel('Position')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
data= np.random.normal(loc=20.0,scale=5.0,size=365)
plt.hist(data,bins=20,color='blue',alpha=0.7)
plt.title("Histogram for Position and Age")
plt.xlabel('Position')
plt.ylabel('Age')
plt.show()
```

```python
data = np.random.normal(loc=20.0, scale=5.0, size=365)
intervals = np.arange(0, 51, 5) 
counts, _ = np.histogram(data, bins=intervals)
plt.bar(intervals[:-1], counts, width=4, color='blue', alpha=0.7)
plt.title("Bar graph for Position and Age")
plt.xlabel('Position')
plt.ylabel('Age')
plt.show()
```

```python
data = np.random.normal(loc=20.0, scale=5.0, size=365)
intervals = np.arange(0, 51, 5)
counts, _ = np.histogram(data, bins=intervals)
percentage = (counts / len(data)) * 100
labels = [f"{intervals[i]}-{intervals[i+1]}" for i in range(len(intervals)-1)]
plt.figure(figsize=(7, 7))
plt.pie(percentage, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie chart for Position and Age')
plt.show()
```
